# Contradiction score/binary/trinary eval

```python
import pandas as pd

df = pd.read_csv('contradiction_score.tsv', sep='\t')
```

```python
df
```

```python
df['average_contradiction'] = df[['0', '1', '2', '3', '4']].mean(axis=1)
```

```python
df.groupby('final_label')['average_contradiction'].mean()
```

```python
from sklearn.metrics import classification_report

threshold = 3

df['categorical_contradiction'] = df.average_contradiction.apply(lambda x: 'W' if x > threshold else 'NW')

print(len(df[(df.final_label != 'U') & (df.categorical_contradiction == df.final_label)]) / len(df[df.final_label != 'U']))

print(classification_report(df.final_label[(df.final_label != 'U')], df[(df.final_label != 'U')].categorical_contradiction))
```

```python
df_binary = pd.read_csv('contradiction_binary.tsv', sep='\t')
df_trinary = pd.read_csv('contradiction_trinary.tsv', sep='\t')

df_binary['categorical_contradiction'] = df_binary[['0', '1', '2', '3', '4']].mode(axis=1)[0].map({0:'NW', 1:'W', -1:'U'})
df_trinary['categorical_contradiction'] = df_trinary[['0', '1', '2', '3', '4']].mode(axis=1)[0].map({0:'NW', 1:'W', -1:'U'})
```

```python
print(classification_report(df_binary.final_label[(df_binary.final_label != 'U')], df_binary[(df_binary.final_label != 'U')].categorical_contradiction))
print(classification_report(df_trinary[(df_trinary.final_label != 'U')].final_label, df_trinary[(df_trinary.final_label != 'U')].categorical_contradiction))
```
